# Finding Longest Common Subsequence of two two pairs of records

```python
# Dynamic programming implementation of LCS problem 
  
# Returns length of LCS for X[0..m-1], Y[0..n-1]  
def lcs(X, Y, m, n): 
    L = [[0 for x in range(n+1)] for x in range(m+1)] 
  
    # Following steps build L[m+1][n+1] in bottom up fashion. Note 
    # that L[i][j] contains length of LCS of X[0..i-1] and Y[0..j-1]  
    for i in range(m+1): 
        for j in range(n+1): 
            if i == 0 or j == 0: 
                L[i][j] = 0
            elif X[i-1] == Y[j-1]: 
                L[i][j] = L[i-1][j-1] + 1
            else: 
                L[i][j] = max(L[i-1][j], L[i][j-1]) 
  
    # Following code is used to print LCS 
    index = L[m][n] 
  
    # Create a character array to store the lcs string 
    lcs = [""] * (index+1) 
    lcs[index] = "" 
  
    # Start from the right-most-bottom-most corner and 
    # one by one store characters in lcs[] 
    i = m 
    j = n 
    while i > 0 and j > 0: 
  
        # If current character in X[] and Y are same, then 
        # current character is part of LCS 
        if X[i-1] == Y[j-1]: 
            lcs[index-1] = X[i-1] 
            i-=1
            j-=1
            index-=1
  
        # If not same, then find the larger of two and 
        # go in the direction of larger value 
        elif L[i-1][j] > L[i][j-1]: 
            i-=1
        else: 
            j-=1
  
    result = "".join(lcs)  
    return result
# Driver program 
#LCS_list = []
#X = list[0]
#Y = list[1]
#m = len(X) 
#n = len(Y) 
#LCS = lcs(X, Y, m, n) 
#print(LCS)
#for i in range(len(list)-2):   
#    X = list[i]
#    Y = list[i+1]
#    m = len(X) 
#    n = len(Y) 
#    LCS = lcs(X, Y, m, n) 
#    LCS_list.append(LCS)
#print(LCS_list)
```

# Finding LCS of even and odd pairs e.g 0 nd 2 pair, 1 nd 3 pair, and so on record,, of 280 DNA Records

```python
# we are going to find LCS of even and odd pairs like 0 and 2, 1 nd 3rd record,and so on,, of 280 records as training data   
even_new_pair_check = True
odd_new_pair_check  = True
LCS_of_even_pairs_of_280_records = []
LCS_of_odd_pairs_of_280_records = []
LCS_dictionary_of_even_odd_pairs = {} # for saving result as a dictionary

for i in range(0,280):
    # creating even pairs,,,even_new_pair_check variable i used to avoid repeatition of record,, because 0 and 2 1st pair,,
    # and next time whenever i varible will incounter 2 again, then i need to bypass this 2 record because 2 record is already used,
    # and same for odd_new_pair_check variable
    if(i%2==0):
        if(even_new_pair_check):
            even_record1 = list[i]
            even_record2 = list[i+2]
            m = len(even_record1) 
            n = len(even_record2)
            even_LCS = lcs(even_record1,even_record2, m, n)
            LCS_of_even_pairs_of_280_records.append(even_LCS)
            even_new_pair_check = False
        else:
            even_new_pair_check = True
    # creating odd pairs
    if(i%2==1):
        if(odd_new_pair_check):
            odd_record1 = list[i]
            odd_record2 = list[i+2]
            m = len(odd_record1) 
            n = len(odd_record2)
            odd_LCS = lcs(odd_record1,odd_record2, m, n)
            LCS_of_odd_pairs_of_280_records.append(odd_LCS)
            odd_new_pair_check = False
        else:
            odd_new_pair_check = True
# saving both lists of LCS of even odd pairs of 280 records into dictionary
LCS_dictionary_of_even_odd_pairs.update({'LCS_of_even_pairs_of_280_records':LCS_of_even_pairs_of_280_records}) # key value pair
LCS_dictionary_of_even_odd_pairs.update({'LCS_of_odd_pairs_of_280_records':LCS_of_odd_pairs_of_280_records}) # key value pair
```

# Finding Longest common substring by making even odd pairs

```python
def get_longest_common_substring(record1,record2):
    # Import pairwise2 module
    from Bio import pairwise2

    # Import format_alignment method
    from Bio.pairwise2 import format_alignment

    # Define two sequences to be aligned
    X = record1
    Y = record2

    # Get a list of the global alignments between the two sequences ACGGGT and ACG
    # No parameters. Identical characters have score of 1, else 0.
    # No gap penalties.
    alignments = pairwise2.align.localxx(X, Y)
    alignment_last = []
    alignmentA = ""
    alignmentB = ""
    if(alignments): #In Python, empty lists evaluate to False and non-empty lists evaluate to True in boolean contexts. 
        alignment_last = alignments[-1]
        alignmentA = alignment_last[0]
        alignmentB = alignment_last[1]
    # Use format_alignment method to format the alignments in the list
    #for a in alignments:
    #    print(format_alignment(*a))

    longest_substring = ""
    current_substring = ""

    for i in range(len(alignmentA)):
        if(alignmentA==alignmentB):
            longest_substring = alignmentA
            break        
     #   print("longest_substring",i,": ",longest_substring)
      #  print("current_substring: ",current_substring,"\n\n")n
        if(alignmentA[i]=='-' or alignmentB[i]=='-'):
            if(len(longest_substring)<len(current_substring)):
                longest_substring = current_substring
            current_substring = ""
        elif(alignmentA[i]==alignmentB[i]):
            current_substring = current_substring + alignmentA[i]
    #print("longest_substring: ",longest_substring)
    return longest_substring
```

```python
all_longest_substring_list = [] 
def even_odd_LCS_generator(x):
    if(len(x)==1):
        return x[0]
    left = even_odd_LCS_generator(x[::2])
    right= even_odd_LCS_generator(x[1::2])
    longest_substring = get_longest_common_substring(left,right)
    all_longest_substring_list.append(longest_substring)
    print("longest_string: ",longest_substring," \n\n")
    return longest_substring
```
